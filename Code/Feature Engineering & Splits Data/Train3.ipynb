Fitur Optimasi

```python
import pandas as pd
import numpy as np

# 1. Load data
input_path = "C:/Python/Template/Data Science/DATATHON_2025/data/off_chain/off_chain_btc.csv"
df = (
    pd.read_csv(input_path, parse_dates=["timestamp"], index_col="timestamp")
      .sort_index()
)

# 2. Core returns & diffs for 'close'
df["close_ret1h"]  = df["close"].pct_change()
df["close_diff1h"] = df["close"].diff()

# 3. Lag features: only close_ret1h at 1h & 3h
for lag in (1, 3):
    df[f"close_ret_lag{lag}h"] = df["close_ret1h"].shift(lag)

# 4. Intrabar features
df["range"]      = df["high"] - df["low"]
df["body"]       = df["close"] - df["open"]
df["body_ratio"] = df["body"] / df["range"].replace(0, np.nan)

# 5. Order‐flow imbalance
df["sell_vol"]    = df["volume"] - df["taker_buy_base_vol"]
df["imbalance"]   = df["taker_buy_base_vol"] - df["sell_vol"]
df["taker_ratio"] = df["taker_buy_base_vol"] / df["volume"].replace(0, np.nan)

# 6. VWAP: only 2h & 24h
for w in (2, 24):
    vol_sum = df["volume"].rolling(w, min_periods=1).sum()
    df[f"vwap_{w}h"] = (
        (df["close"] * df["volume"]).rolling(w, min_periods=1).sum()
        / vol_sum.replace(0, np.nan)
    )

# 7. Rolling Z‐score 24h: only on key vars
for col in ("close", "volume", "num_trades"):
    mu    = df[col].rolling(24, min_periods=1).mean()
    sigma = df[col].rolling(24, min_periods=1).std()
    df[f"{col}_z_24h"] = (df[col] - mu) / sigma.replace(0, np.nan)

# 8. Calendar cyclical: hour & day‐of‐week
idx = df.index
hrs = idx.hour
dow = idx.dayofweek
df["hour_sin"] = np.sin(2 * np.pi * hrs / 24)
df["hour_cos"] = np.cos(2 * np.pi * hrs / 24)
df["dow_sin"]  = np.sin(2 * np.pi * dow / 7)
df["dow_cos"]  = np.cos(2 * np.pi * dow / 7)

# 9. Session flags (UTC)
df["session_asia"]   = ((hrs >= 0)  & (hrs <  8)).astype(int)
df["session_europe"] = ((hrs >= 8)  & (hrs < 16)).astype(int)
df["session_us"]     = ((hrs >= 16) & (hrs < 24)).astype(int)

# 10. Rolling stats windows 6h & 24h
for w in (6, 24):
    df[f"ret_mean_{w}h"] = df["close_ret1h"].rolling(w, min_periods=1).mean()
    df[f"ret_std_{w}h"]  = df["close_ret1h"].rolling(w, min_periods=1).std()
    df[f"vol_mean_{w}h"] = df["volume"].rolling(w, min_periods=1).mean()
    df[f"vol_std_{w}h"]  = df["volume"].rolling(w, min_periods=1).std()

# 11. EMA, MACD, RSI & Bollinger Bands
df["ema_12"]      = df["close"].ewm(span=12, adjust=False).mean()
df["ema_26"]      = df["close"].ewm(span=26, adjust=False).mean()
df["macd"]        = df["ema_12"] - df["ema_26"]
df["macd_signal"] = df["macd"].ewm(span=9, adjust=False).mean()

delta     = df["close"].diff()
gain      = delta.clip(lower=0)
loss      = -delta.clip(upper=0)
avg_gain  = gain.rolling(14, min_periods=1).mean()
avg_loss  = loss.rolling(14, min_periods=1).mean()
rs        = avg_gain / avg_loss.replace(0, np.nan)
df["rsi_14"] = 100 - (100 / (1 + rs))

df["bb_mid"]   = df["close"].rolling(20, min_periods=1).mean()
df["bb_std"]   = df["close"].rolling(20, min_periods=1).std()
df["bb_upper"] = df["bb_mid"] + 2 * df["bb_std"]
df["bb_lower"] = df["bb_mid"] - 2 * df["bb_std"]

# 12. Fill missing di baris awal (backward-fill), lalu fill forward untuk safety
df.fillna(method="bfill", inplace=True)
df.fillna(method="ffill", inplace=True)

# 13. Simpan hasil
output_path = "C:/Python/Template/Data Science/DATATHON_2025/data/off_chain/off_chain_btc2.csv"
df.to_csv(output_path)

print(f"✅ Fitur diperbarui dengan backfill & saved to:\n   {output_path}")
```

```python
import pandas as pd
import numpy as np

# 1. Load data (timestamp jadi index)
input_path = "C:/Python/Template/Data Science/DATATHON_2025/data/off_chain/off_chain_btc.csv"
df = pd.read_csv(
    input_path,
    parse_dates=["timestamp"],
    index_col="timestamp"
)

# 2. Definisikan horizon untuk hourly (jam) dan daily (hari)
hourly_horizons = [1, 2, 3, 6, 12]      # dalam satuan jam
daily_horizons  = [1, 3, 7, 15, 30]     # dalam satuan hari

# 3. Buat kolom label harga untuk tiap horizon
for h in hourly_horizons:
    df[f"label_{h}h"] = df["close"].shift(-h)

for d in daily_horizons:
    shift_hours = d * 24
    df[f"label_{d}d"] = df["close"].shift(-shift_hours)

# 4. Buang baris yang label-nya NaN (akhir dataset)
label_cols = [f"label_{h}h" for h in hourly_horizons] \
           + [f"label_{d}d" for d in daily_horizons]

df_clean = df.dropna(subset=label_cols)

# 5. Simpan hasil
output_path = "C:/Python/Template/Data Science/DATATHON_2025/data/off_chain/off_chain_btc4.csv"
df_clean.to_csv(output_path, index=True)

print(f"✅ Label harga multi-horizon ({len(label_cols)} kolom) dibuat dengan prefix 'label_' dan disimpan di\n   {output_path}")
```

Optimasi Fitur

```python
import pandas as pd
import numpy as np
import os

# Paths
input_path  = "C:/Python/Template/Data Science/DATATHON_2025/data/on_chain/on_chain_btc2.csv"
output_path = os.path.join(os.path.dirname(input_path), "on_chain_btc3.csv")

# Load & sort
df = (
    pd.read_csv(input_path, parse_dates=["timestamp"], index_col="timestamp")
      .sort_index()
)

# 1. Daily returns (1d = 24h)
raw_vars = [
    "HashRate","AdrActCnt","TxCnt","TxTfrValAdjUSD",
    "FeeMeanUSD","FeeTotUSD","CapMVRVCur","NVTAdj",
    "SplyAct1d","SplyAct7d","SplyAct30d","IssContPctDay",
    "FlowInExUSD","FlowOutExUSD"
]
for col in raw_vars:
    df[f"{col}_ret1d"] = df[col].pct_change(24)

# 2. Rolling std of 1d returns (volatility)
for col in raw_vars:
    df[f"{col}_std1d"] = df[f"{col}_ret1d"].rolling(24, min_periods=1).std()

# 3. Lag-1d for key predictors
for col in ["CapMVRVCur","NVTAdj","SplyAct1d"]:
    df[f"{col}_lag1d"] = df[col].shift(24)

# 4. Z-score 1d for valuation metrics
for col in ["CapMVRVCur","NVTAdj"]:
    mu    = df[col].rolling(24, min_periods=1).mean()
    sigma = df[col].rolling(24, min_periods=1).std().replace(0, np.nan)
    df[f"{col}_z1d"] = (df[col] - mu) / sigma

# 5. Composite & ratio features
df["HashRate_ratio"] = df["HashRate"] / df["HashRate30d"].replace(0, np.nan)
df["FlowNetUSD"]     = df["FlowInExUSD"] - df["FlowOutExUSD"]
df["FlowRatio"]      = df["FlowInExUSD"] / df["FlowOutExUSD"].replace(0, np.nan)
df["SplyAct_spread"] = df["SplyAct7d"] - df["SplyAct1d"]

# 6. Fill gaps
df.ffill(inplace=True)
df.bfill(inplace=True)

# 7. Save reduced-feature dataset
df.to_csv(output_path)
```

```python
import pandas as pd
import numpy as np
import os

# Paths
input_path  = "C:/Python/Template/Data Science/DATATHON_2025/data/macro/macro3/macro_btc2.csv"
output_path = os.path.join(os.path.dirname(input_path), "macro_btc3.csv")

# Load & sort
df = pd.read_csv(input_path, parse_dates=["timestamp"], index_col="timestamp").sort_index()

# 1. Impute missing weekends / gaps
df.ffill(inplace=True)
df.bfill(inplace=True)

# 2. Daily & weekly returns (1d = 24h, 7d = 168h)
raw_vars = df.columns.tolist()
for col in raw_vars:
    df[f"{col}_ret1d"] = df[col].pct_change(24)
    df[f"{col}_ret7d"] = df[col].pct_change(24 * 7)

# 3. Rolling volatility (std) on 1d & 7d returns
for col in raw_vars:
    df[f"{col}_vol1d"] = df[f"{col}_ret1d"].rolling(24, min_periods=1).std()
    df[f"{col}_vol7d"] = df[f"{col}_ret7d"].rolling(24 * 7, min_periods=1).std()

# 4. Lag features (1d, 7d) for key predictors
lag_vars = ["vix_index", "fed_funds_rate", "cpi_us"]
for col in lag_vars:
    df[f"{col}_lag1d"] = df[col].shift(24)
    df[f"{col}_lag7d"] = df[col].shift(24 * 7)

# 5. Z‐score 21d for Fed balance sheet (21d = 504h)
window = 24 * 21
mu    = df["fed_balance_sheet"].rolling(window, min_periods=1).mean()
sigma = df["fed_balance_sheet"].rolling(window, min_periods=1).std().replace(0, np.nan)
df["fed_balance_sheet_z21d"] = (df["fed_balance_sheet"] - mu) / sigma

# 6. Final fill for any new gaps
df.ffill(inplace=True)
df.bfill(inplace=True)

# Save reduced-feature macro dataset
df.to_csv(output_path)
```

Penggabungan dan split data

```python
import pandas as pd
import os

# Paths
off_chain_path = "C:/Python/Template/Data Science/DATATHON_2025/data/off_chain/off_chain_btc4.csv"
on_chain_path  = "C:/Python/Template/Data Science/DATATHON_2025/data/on_chain/on_chain_btc2.csv"
macro_path     = "C:/Python/Template/Data Science/DATATHON_2025/data/macro/macro3/macro_btc2.csv"
# C:\Python\Template\Data Science\DATATHON_2025\data\macro\macro3\macro_btc3.csv
output_dir     = r"C:/Python/Template/Data Science/DATATHON_2025/data/fix2"

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Load dataframes
df_off   = pd.read_csv(off_chain_path, parse_dates=["timestamp"], index_col="timestamp")
df_on    = pd.read_csv(on_chain_path,  parse_dates=["timestamp"], index_col="timestamp")
df_macro = pd.read_csv(macro_path,     parse_dates=["timestamp"], index_col="timestamp")

# Merge in order: macro, on-chain, off-chain
df_merged = df_macro.join(df_on, how="inner").join(df_off, how="inner")

# Identify macro-only columns to drop with >90% zeros
threshold = 0.30
macro_cols = df_macro.columns
zero_ratio_macro = (df_merged[macro_cols] == 0).sum() / len(df_merged)
drop_macro = zero_ratio_macro[zero_ratio_macro > threshold].index.tolist()

print("Dropped macro variables (>{:.0%} zeros):".format(threshold))
for col in drop_macro:
    print(f" - {col}")

# Drop only those macro columns
df_merged.drop(columns=drop_macro, inplace=True)

# Split into train/val/test
n = len(df_merged)
train_end = int(0.95 * n)
val_end   = int(0.98 * n)

df_train = df_merged.iloc[:train_end]
df_val   = df_merged.iloc[train_end:val_end]
df_test  = df_merged.iloc[val_end:]

# Save splits
df_train.to_csv(os.path.join(output_dir, "train.csv"), index=True)
df_val.to_csv(os.path.join(output_dir, "val.csv"),   index=True)
df_test.to_csv(os.path.join(output_dir, "test.csv"),  index=True)
```
